# Problem Statement

Air travel is an essential aspect of modern life, with millions of people relying on it for business, leisure, and personal connections. The fluctuating nature of flight prices, influenced by factors such as supply, demand, seasonality, and competition, often leaves travelers overwhelmed when trying to find the best deals on airfare. At the same time, airlines and travel companies are constantly seeking to optimize their pricing strategies to maximize revenues and maintain customer satisfaction. Despite the abundance of flight data available, accurately forecasting flight prices remains a challenge for both travelers and industry stakeholders. Our goal is to develop a flight price predictor that leverages big data techniques and machine learning algorithms to accurately forecast flight prices, enabling travelers to make informed decisions and save money on air travel. By providing essential information, such as departure date and time, arrival time, source and destination, number of stops, and airline name, our solution aims to empower users with the insights they need while also assisting airlines and travel companies in refining their pricing strategies.

### **4.B. Analyzing average fare by each month:**

```python
# Extracting each month, date and year from the column "flightDate" and dropping "flightDate":

flights = flights.withColumn("flight_day", dayofmonth("flightDate"))
flights = flights.withColumn("flight_year", year("flightDate"))
flights = flights.withColumn("flight_month", date_format("flightDate", "MMMM"))
```

```python
flights = flights.drop('flightDate')
```

```python
flights.columns
```

```python
averageFare = flights.groupBy('flight_month').agg(avg('baseFare').alias('averageFare')).orderBy('flight_month')
averageFare_pd = averageFare.toPandas()

colors_1 = ['#00224e', '#123570', '#3b496c', '#575d6d', '#707173', '#8a8678', '#a59c74', '#c3b369']

plt.figure(figsize=(10,5))

plt.bar(averageFare_pd['flight_month'], averageFare_pd['averageFare'], color=colors_1)
plt.xlabel('Month')
plt.ylabel('Average Fare')
plt.title('Average Fare by Month')

plt.show()
```

It appears that data is present for the months of **April, May, June, July, August, September, October, and November**. The graph illustrates that the peak average base fare price occurs in the month of **June**, followed closely by **July**. A possible explanation could be that as summer months are popular for vacations and travel, airlines often raise prices to capitalize on the surge in demand from travelers.

### **4.C. Analyzing months with the most or least flights:**

```python
tickets_by_date = flights.groupBy('flight_month').agg(count('*').alias('num_tickets')).orderBy('flight_month', ascending=True)

most_tickets_dates = tickets_by_date.orderBy('num_tickets', ascending=False).limit(10)
most_tickets_dates.show()
```

```python
tickets_by_date_2 = tickets_by_date.toPandas()

plt.figure(figsize=(8,5))

plt.plot(tickets_by_date_2['flight_month'], tickets_by_date_2['num_tickets'], color="midnightblue")
plt.xlabel('Flight Month')
plt.ylabel('Number of purchasable tickets')
plt.title('Distribution of tickets by Month')

plt.show()
```

The graph above indicates that the months of **August** and **September** had the highest number of ticket sales, with over 16 million tickets sold each. These results aligns with the first graph plotted above when analyzing average fare by each month, suggesting that airlines may have increased flight frequencies during the summer months to accommodate the high demand for air travel. As a result, we can conclude that airlines not only charge higher fares during peak travel periods but also increase the number of flights available to meet the growing demand from travelers.

### **4.D. Analyzing relationship between airlines and base fare:**

```python
airline_fares = flights.groupBy('airline_name').agg(avg('baseFare').alias('averageFare'))
airline_fares_pd = airline_fares.toPandas()

fig, ax = plt.subplots(figsize=(20, 6))

max_fare = airline_fares_pd['averageFare'].max()
max_fare_airline = airline_fares_pd.loc[airline_fares_pd['averageFare'] == max_fare, 'airline_name'].iloc[0]

bars = ax.bar(airline_fares_pd['airline_name'], airline_fares_pd['averageFare'], color='blue')

for i, bar in enumerate(bars):
    if airline_fares_pd.loc[i, 'airline_name'] == max_fare_airline:
        bar.set_color('red')
    
ax.set_xlabel('Airline Name', fontsize=12)
ax.set_ylabel('Average Fare', fontsize=12)
ax.set_title('Average Ticket Price by Airline', fontsize=14)
plt.xticks(rotation=90, fontsize=8)
plt.xticks(range(0, len(airline_fares_pd)), airline_fares_pd['airline_name'], rotation=90, fontsize=8)

plt.show()
```

The graph above demonstrates that airlines frequently collaborate and form alliances, such as Star Alliance, resulting in unique combinations of different airlines working together. We can see that **JetBlue in conjunction with American Airlines** have the highest base fare, however, to better understand the impact of a single airline company on fare, we'll need to delve deeper into the dynamics of airline collaborations and acquire more domain knowledge about the industry.
